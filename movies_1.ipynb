```python
import pandas as pd
import seaborn as sns
import numpy as np 
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
import matplotlib.pyplot as plt
%matplotlib inline 
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
# importing the data 

df_movies = pd.read_csv('/Users/darshanpatel/Desktop/python/movies.csv')

# let's look at the data 

df_movies.head()
```

```python
# data cleaning 
# checking null values 

for col in df_movies.columns:
    pct_missing = np.mean(df_movies[col].isnull())
    print('{}  - {}%' .format(col, pct_missing))
```

```python
#droping nalue values 

df_movies.dropna(inplace=True)
```

```python
# checking null values again  

for col in df_movies.columns:
    pct_missing = np.mean(df_movies[col].isnull())
    print('{}  - {}%' .format(col, pct_missing))
```

```python

df_movies.dtypes
```

```python
# changing data types 

df_movies['budget'] = df_movies['budget'].astype('int64')
df_movies['gross'] = df_movies['gross'].astype('int64')
df_movies['votes'] = df_movies['votes'].astype('int64')
df_movies['runtime'] = df_movies['runtime'].astype('int64')

df_movies.dtypes
```

```python

df_movies['year_fix'] = df_movies['released'].astype('str').str.extract('(\d{4})')

df_movies.head()
```

```python
df_movies.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)

df_movies.head()
```

```python
# droping duplicate 

df_movies.drop_duplicates()

df_movies.head()
```

```python
# scatter plot with budget  vs gross revenue 

plt.scatter(x=df_movies['budget'], y=df_movies['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# plot budget vs gross seaborn 

sns.regplot(x='budget' , y='gross', data=df_movies, scatter_kws={"color" :"green"} , line_kws={"color" :"red"} )
```

```python
 # lets start looking at correlation
    
df_movies.corr() 
```

```python
correlation_matrix = df_movies.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation_matrix')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')

plt.show()

```

```python
df_movies_numerized = df_movies.copy()

for col_name in df_movies.select_dtypes(include=['object']):
    df_movies_numerized[col_name] = df_movies[col_name].astype('category')
    df_movies_numerized[col_name] = df_movies_numerized[col_name].cat.codes
      
```

```python
df_movies_numerized

df_movies_numerized.head()
```

```python
df_movies_numerized.corr() 
```

```python
correlation_matrix = df_movies_numerized.corr()

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation_matrix')

plt.xlabel('Movies Features')

plt.ylabel('Movies Features')

plt.show()
```

```python
correlation_mat = df_movies_numerized.corr()

corr = correlation_mat.unstack()

corr
```

```python
sor = corr.sort_values()
sor
```

```python
high_corr = sor[(sor) > 0.5]
high_corr
```
